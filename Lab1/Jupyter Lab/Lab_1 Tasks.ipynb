<center><h4>{Lists}</h4></center>

```python
mylist = []
mylist.append(1) 
mylist.append(2) 
mylist.append(3)

print(mylist[0])
print (mylist[-1])
(mylist[1:3])
print("")

names = ["John", 3234, 2342, 3323, "Eric", 234, "Jessica",
734978234, "Lois", 2384]
print("Number of names in list: {}".format(len(names)))

new_names = []
for n in names:
    if isinstance(n, str):
        new_names.append(n)

new_names.sort()
print("Cleaned-up number of names in list: {}".format(len (new_names)))

for i, n in enumerate (new_names):
    print("{}. {}".format(i+1, n))
```
